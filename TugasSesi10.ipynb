```python
import numpy as np
import pandas as pd

jmlData = 1000
np.random.seed(123)

usia = np.random.randint(15, 66, size=jmlData)
jenis_kelamin = np.random.choice(['pria', 'wanita'], size=jmlData)
frekuensi_olahraga = np.random.randint(0, 8, size=jmlData)  
durasi_olahraga = np.random.randint(10, 121, size=jmlData)  
tingkat_kebugaran = np.random.choice(['Rendah', 'Sedang', 'Tinggi'], size=jmlData)


df_kebugaran = pd.DataFrame({
    'Usia': usia,
    'JenisKelamin': jenis_kelamin,
    'FrekuensiOlahragaPerMinggu': frekuensi_olahraga,
    'DurasiOlahragaPerSesi': frekuensi_olahraga,
    'TingkatKebugaran': tingkat_kebugaran
})

print(df_kebugaran)
print("\nInformasi DataFrame :")
print(df_kebugaran.info())
print("\n5 Data Pertama :")
print(df_kebugaran.head())

df_kebugaran.to_csv("dataKebugaran.csv", index=False)

print(f"\nUsia Termuda: {df_kebugaran['Usia'].min()} tahun")
print(f"Usia  Tertua: {df_kebugaran['Usia'].max()} tahun")

pria = df_kebugaran[df_kebugaran['JenisKelamin'] == 'Pria']
print(f"\nJumlah peserta pria: {pria.shape[0]}")
print(f"Rata-rata durasi olahraga pria : {pria['DurasiOlahragaPerSesi'].mean():.2f} menit")

wanita = df_kebugaran[df_kebugaran['JenisKelamin'] == 'Wanita']
print(f"\nJumlah peserta wanita: {wanita.shape[0]}")
print(f"Rata-rata frekuensi olahraga wanita : {wanita['FrekuensiOlahragaPerMinggu'].mean():.2f} kali/minggu")
```
